# EDA

Converting dates to datetime format

```python
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])
df_transactions['TransactionDate'] = pd.to_datetime(df_transactions['TransactionDate'])
```

# Region distribution

```python
region_distribution = df_customers["Region"].value_counts()
```

```python
plt.subplot(2, 2, 1)
sns.countplot(data=df_customers, y="Region", order=region_distribution.index, palette="viridis")
plt.title("Customer Distribution by Region")
plt.xlabel("Count")
plt.ylabel("Region")
```

# Category distribution

```python
category_distribution = df_products["Category"].value_counts()
```

```python
plt.subplot(2, 2, 2)
sns.countplot(data=df_products, y="Category", order=category_distribution.index, palette="magma")
plt.title("Product Distribution by Category")
plt.xlabel("Count")
plt.ylabel("Category")
```

# Total transactions by region

```python
transactions_by_region = df_transactions.merge(df_customers,on='CustomerID')
transactions_region_value = transactions_by_region.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
```

```python
plt.subplot(2, 2, 3)
transactions_region_value.plot(kind="bar", color="teal")
plt.title("Total Transaction Value by Region")
plt.xlabel("Region")
plt.ylabel("Total Value (USD)")
```

# Top 10 products by total sales

```python
transactions_with_products = df_transactions.merge(df_products, on="ProductID")
top_products = (
    transactions_with_products.groupby("ProductName")["TotalValue"].sum().nlargest(10)
)
```

```python
plt.subplot(2, 2, 4)
top_products.plot(kind="barh", color="orange")
plt.title("Top 10 Products by Total Sales Value")
plt.xlabel("Total Sales Value (USD)")
plt.ylabel("Product Name")

plt.tight_layout()
plt.show()
```

# Average product price and transaction value

```python

avg_product_price = df_products['Price'].mean()
avg_transaction_value = df_transactions['TotalValue'].mean()


pricing_insights = pd.DataFrame({
    'Category': ['Average Product Price', 'Average Transaction Value'],
    'Value': [avg_product_price, avg_transaction_value]
})

plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=pricing_insights, palette='coolwarm')
plt.title('Transaction and Product Pricing Insights')
plt.ylabel('Average Value (USD)')
plt.xlabel('Category')
plt.ylim(0, max(avg_product_price, avg_transaction_value) + 100)
plt.text(0, avg_product_price + 30, f"${avg_product_price:.2f}", ha='center', color='black', fontsize=12)
plt.text(1, avg_transaction_value + 30, f"${avg_transaction_value:.2f}", ha='center', color='black', fontsize=12)
plt.show()
```

# Sales Trends Over Time

```python
df_transactions['TransactionDate'] = pd.to_datetime(df_transactions['TransactionDate'])

monthly_sales = df_transactions.groupby(df_transactions['TransactionDate'].dt.to_period('M'))['TotalValue'].sum().reset_index()
monthly_sales['TransactionDate'] = monthly_sales['TransactionDate'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
sns.lineplot(x='TransactionDate', y='TotalValue', data=monthly_sales, marker='o', color='blue')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

# Top-Selling Products

```python
product_sales = df_transactions.groupby('ProductID').agg(
    TotalQuantity=('Quantity', 'sum'),
    TotalRevenue=('TotalValue', 'sum')
).reset_index()


product_sales = product_sales.merge(df_products, on='ProductID')
top_products_revenue = product_sales.nlargest(10, 'TotalRevenue')

plt.figure(figsize=(12, 6))
sns.barplot(y='ProductName', x='TotalRevenue', data=top_products_revenue, palette='viridis')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Total Revenue (USD)')
plt.ylabel('Product Name')
plt.show()
```

# Customer Signup Trends

```python
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])

signup_trends = df_customers.groupby(df_customers['SignupDate'].dt.to_period('M')).size().reset_index(name='Count')
signup_trends['SignupDate'] = signup_trends['SignupDate'].dt.to_timestamp()

plt.figure(figsize=(10, 6))
sns.lineplot(x='SignupDate', y='Count', data=signup_trends, marker='o', color='green')
plt.title('Customer Signup Trends')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.xticks(rotation=45)
plt.grid(alpha=0.3)
plt.show()
```

# Regional Sales Performance

```python
transactions_with_customers = df_transactions.merge(df_customers, on='CustomerID')

regional_sales = transactions_with_customers.groupby('Region')['TotalValue'].sum().reset_index()


plt.figure(figsize=(8, 5))
sns.barplot(x='Region', y='TotalValue', data=regional_sales, palette='coolwarm')
plt.title('Regional Sales Performance')
plt.xlabel('Region')
plt.ylabel('Total Sales (USD)')
plt.show()
```

# Category wise revenue distribution

```python
transactions_with_products = df_transactions.merge(df_products, on='ProductID')
category_revenue = transactions_with_products.groupby('Category')['TotalValue'].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='TotalValue', data=category_revenue, palette='Spectral')
plt.title('Category-Wise Revenue Distribution')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```
